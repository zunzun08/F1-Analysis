```python
import pandas as pd
import fastf1 as ff1
```

```python
df = pd.read_csv("/Users/cristianzuniga/Documents/GitHub/F1-Analysis/Ergast Data/results.csv")
df_race = pd.read_csv("/Users/cristianzuniga/Documents/GitHub/F1-Analysis/Ergast Data/races.csv")
```

```python
df_race = df_race[df_race["year"] >= 2018].reset_index(drop=True)
```

```python
df_race.head()
```

```python

df_race[["Temperature (C)", "Humidity", "WindSpeed", "TrackTemp", "Rain"]] = ''
for i in range(df_race.shape[0]):
    weekend = ff1.get_session(df_race["year"][i], df_race["name"][i], "R")
    weekend.load(telemetry=False, weather=True)
    try:
        weather_df = weekend.weather_data
        df_race["Temperature (C)"].loc[i] = weather_df["AirTemp"].mean()
        df_race["Humidity"].loc[i] = weather_df["Humidity"].mean()
        df_race["WindSpeed"].loc[i] = weather_df["WindSpeed"].mean()
        df_race["TrackTemp"].loc[i] = weather_df["TrackTemp"].mean()
        if weather_df["Rainfall"].any() == True:
            df_race["Rain"].loc[i] = True
        else:
            df_race["Rain"].loc[i] = False
    except DataNotLoadedError as err:
        df_race["Temperature (C)"].loc[i] = pd.NA
        df_race["Humidity"].loc[i] = pd.NA
        df_race["WindSpeed"].loc[i] = pd.NA
        df_race["TrackTemp"].loc[i] = pd.NA
        df_race["Rain"].loc[i] = pd.NA
```

```python
df_race = df_race[~(df_race["Rain"] == "")]
df_race = df_race[["raceId", "year", "name", "Temperature (C)", "Humidity", "WindSpeed", "TrackTemp", "Rain"]]
df_race
```

```python
def heat_index(t,h):
    coef = -42.379 + (2.04901523 * t)+(10.14333127 * h)-(0.22475541 * t * h)-(6.83783e-3 * t**2)-(5.481717e-2 * h**2 ) + (1.22874e-3 * t**2 * h) + (8.5282e-4 * t * h**2) - (1.99e-6 * t**2 * h**2)
    if coef >= 80 and coef <=90:
        return "Caution"
    elif coef >= 91 and coef <= 103:
        return "Extreme Caution"
    elif coef >= 104 and coef <= 124:
        return "Danger"
    elif coef >= 125:
        return "Extreme Danger" 
    else:
        return "Fine"
df_race["Temperature (F)"] = (df_race["Temperature (C)"] * 9/5) + 32
df_race["Heat Index"] = ''
for i in range(df_race["Temperature (F)"].shape[0]):
    temp = df_race["Temperature (F)"].loc[i]
    humidity = df_race["Humidity"].loc[i]
    df_race["Heat Index"].loc[i] = heat_index(temp, humidity)
```

```python
df = df.merge(df_race, how="inner", on="raceId")
```

```python
df.to_csv("races (with weather 2018-).csv")
```
